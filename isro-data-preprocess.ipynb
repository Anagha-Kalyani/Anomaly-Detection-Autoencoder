```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = ["DTG-1_PCH_ANALOG_RAT (Analog)", "DTG-1_ROL_ANALOG_RAT (Analog)",
                         "DTG-1_PCH_FINE_RATE (Analog)", "DTG-1_ROL_FINE_RATE (Analog)",
                         "DTG-2_YAW_ANALOG_RAT (Analog)", "DTG-2_PCH_ANALOG_RAT (Analog)",
                         "DTG-2_YAW_FINE_RATE (Analog)", "DTG-2_PCH_FINE_RATE (Analog)",
                         "DTG-1_TH_TEMP (Analog)", "DTG-1_ELECTRONICS_TH (Analog)",
                         "DTG-2_TH_TEMP (Analog)", "DTG-2_ELECTRONICS_TH (Analog)"]

scaler = MinMaxScaler()

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
```

```python
def derive_month_day(row):
    year = row['Year']
    day_of_year = row['Day of year']
    
    # Create a datetime object for the given year and day of the year
    date = pd.to_datetime(f'{year}-{day_of_year}', format='%Y-%j')
    
    
    # Extract month and day from the datetime object
    row['Month'] = date.month
    row['Day'] = date.day
    
    return row

# Apply the function to the DataFrame
df = df.apply(derive_month_day, axis=1)

# Drop the original 'Year' and 'Day of the year' columns if needed
df = df.drop(columns=['Day of year'])

print(df.head(5))
```

```python
df['Datetime'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Millisecond', 'Microsecond']])
```

```python
df=df.drop(['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Millisecond', 'Microsecond'], axis=1)
```
